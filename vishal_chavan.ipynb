## Defining a general function to plot the dataset

```python
def plotData(x, y, plotType:str):
    plt.title("{}".format(plotType))
    if plotType == "Scatter":
        plt.scatter(x, y, color = 'green')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()

    if plotType == "Line":
        plt.plot(x, y, color = 'red')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()

    if plotType == "Bar":
        plt.bar(x, y)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()

   
        
        
    
```

### 1. Scatter plot

```python
plotData(x_points, y_points, "Scatter")
```

### 2. Line Plot

```python
plotData(x_points, y_points, "Line")
```

```python
plotData(x_points, y_points, "Bar")
```

```python
x_val = []
y_val = []

for i in data['input']:
    if i == np.nan:
        continue
    x_val.append(i)

for j in data['output']:
    if j == np.nan:
        continue
    y_val.append(j)


x_val = np.array(x_val)
x_val = x_val.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_val, y_val, test_size=0.2)
```
